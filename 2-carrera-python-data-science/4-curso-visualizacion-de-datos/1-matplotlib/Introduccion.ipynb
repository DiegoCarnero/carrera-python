```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
x = np.arange(5)
y = np.array([2.0, 2.5, 1.5, 3.0, 4.5])
z = y + 1 
w = np.log(y)
```

```python
plt.plot(x,y, '*', label = 'y')
plt.plot(x,z, 'g-', label = 'y')
plt.plot(x,w, '--', label = 'y')
plt.title('Grafico')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
```
