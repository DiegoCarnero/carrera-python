```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('mpg')
data.head()
```

```python
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=2, hue='displacement', style='cylinders');
```

```python
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=2, hue='origin', size='cylinders');
```

```python
hp_mean = data.groupby('cylinders')['horsepower'].mean()
hp_mean
```

```python
sns.relplot(x = 'cylinders', y = 'horsepower', data = data, height=7, kind = 'line', hue = 'origin', ci = 'sd');
```
