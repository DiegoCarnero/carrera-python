```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
g = sns.FacetGrid(data = data, col = 'sex')
g.map(sns.histplot, 'age');
```

```python
g = sns.FacetGrid(data = data, col = 'pclass', hue = 'sex')
g.map(sns.scatterplot, 'age', 'fare')
g.add_legend()
for ax in g.axes.flat:
    ax.axhline(y = 300)
```

```python
g = sns.PairGrid(data = data, vars = ['pclass', 'age', 'fare'], hue = 'sex')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
sns.pairplot(data = data, vars = ['pclass', 'age', 'fare'], hue = 'sex', kind = 'scatter')
```
