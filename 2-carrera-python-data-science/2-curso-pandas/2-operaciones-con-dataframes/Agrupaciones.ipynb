```python
agrup = df.groupby('condition').agg({
    'price' : [np.mean,'min','max'],
    'bathrooms' : 'median',
    'waterfront' : np.max,
    'floors' : 'min'
})
agrup
```

```python
def media_mayores_1millon(x):
    return np.mean(x[x > 1_000_000])
```

```python
df[['price','condition']].groupby('condition')['price'].apply(media_mayores_1millon)
```
