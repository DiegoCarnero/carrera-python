Ejercicio 4

- Convierte la columna 'date' en un tipo fecha.
- Genera 3 nuevas columnas fruto de extraer el año, mes y dia.
- ¿Cual es el precio medio por mes?
- Crea una nueva columna fruto de sumar 20 días a la columna fecha.

```python
df = pd.read_csv('data/house.csv')
df['date']
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['anho']=df['date'].dt.year
df['mes']=df['date'].dt.month
df['dia']=df['date'].dt.day
df.sample(1)
```

```python
df.set_index(df['date']).resample('1M')['price'].mean()
```

```python
df['mes_mas_20'] = df['date'] + pd.Timedelta('20 days')
df[['date', 'mes_mas_20']].sample(5)
```
